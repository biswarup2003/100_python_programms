```python
import os
import shutil
import pandas as pd

def copy_text_file_to_folders(text_file_path, parent_folder, output_folder):
    """
    Copies the specified text file to each subfolder inside the parent folder and saves it with the same name as the subfolder.
    """
    # Ensure the text file exists
    if not os.path.exists(text_file_path):
        print(f"Error: The text file at {text_file_path} does not exist.")
        return
    
    # Ensure the parent folder exists
    if not os.path.isdir(parent_folder):
        print(f"Error: The parent folder at {parent_folder} does not exist.")
        return

    # Ensure the output folder exists, if not create it
    if not os.path.exists(output_folder):
        print(f"Output folder does not exist. Creating it: {output_folder}")
        os.makedirs(output_folder)

    # Get the list of subfolders in the parent folder
    subfolders = [f for f in os.listdir(parent_folder) if os.path.isdir(os.path.join(parent_folder, f))]
    if not subfolders:
        print("No subfolders found in the specified location.")
        return

    # Copy the text file into each folder and rename it based on the folder's name
    for folder in subfolders:
        folder_path = os.path.join(parent_folder, folder)
        new_file_path = os.path.join(output_folder, f"{folder}.txt")
        
        try:
            shutil.copy(text_file_path, new_file_path)
            print(f"Copied text file to: {new_file_path}")
        except Exception as e:
            print(f"Failed to copy to {folder}: {e}")

def process_files(input_directory, output_directory):
    """
    Processes the files inside each subfolder by reading a random .txt file and updating it with additional columns.
    The updated data is saved back into the corresponding output file.
    """
    # List all subfolders in the input directory
    subfolders = [f for f in os.listdir(input_directory) if os.path.isdir(os.path.join(input_directory, f))]
    
    # Iterate through each subfolder
    for subfolder in subfolders:
        # Construct the full path to the corresponding output file
        output_file_path = os.path.join(output_directory, subfolder + '.txt')  # Output file has the same name as the subfolder
        
        # Check if the corresponding output file exists
        if os.path.exists(output_file_path):
            print(f"Processing files for subfolder: {subfolder}")
            
            # Find the .txt file in the input subfolder
            input_subfolder_path = os.path.join(input_directory, subfolder)
            input_files = [f for f in os.listdir(input_subfolder_path) if f.endswith('.txt')]  # List all .txt files
            
            if input_files:
                input_file_path = os.path.join(input_subfolder_path, input_files[0])  # Pick the first .txt file found
                print(f"Found input file: {input_file_path}")
                
                try:
                    # Read the input data (random .txt file from the subfolder)
                    df = pd.read_csv(input_file_path, sep='|')

                    # Read the header template from the corresponding output file
                    temp_df = pd.read_csv(output_file_path, sep='|')

                    # Remove the last column from the header template
                    temp_df = temp_df.iloc[:, :-1]

                    # Create a new DataFrame with the same number of rows as the data
                    temp_df = pd.DataFrame(index=range(len(df)), columns=temp_df.columns)

                    # Add 'Batch' and 'ImageType' columns
                    temp_df['Batch'] = df['FS_DIGITAL_FILM_NBR'].values
                    temp_df['ImageType'] = 'Baptism'

                    # Fill NaN in 'FS_IMAGE_NBR' column with 0 and convert to integer
                    df['FS_IMAGE_NBR'] = df['FS_IMAGE_NBR'].fillna(0).astype(int)

                    # Generate image numbers
                    image_no = temp_df['Batch'].astype(str) + "_" + df['FS_IMAGE_NBR'].apply(lambda x: f"{int(x):05d}" if pd.notna(x) else '00000')
                    temp_df['Image'] = image_no

                    # Overwrite the existing output file with the updated header template
                    temp_df.to_csv(output_file_path, sep='|', index=False)
                    
                    print(f"File updated successfully for subfolder: {subfolder}")

                except Exception as e:
                    print(f"An error occurred while processing {subfolder}: {e}")
            else:
                print(f"No .txt file found in subfolder: {subfolder}")
        else:
            print(f"Corresponding output file for {subfolder} not found.")

# Main execution:

# Define the paths
text_file_path = r'C:\Users\3452268\Documents\Biswarup\test\NA221\headertemplate.txt'  # Template file path
input_directory = r'C:\Users\3452268\Documents\Biswarup\test\NA221\rework'  # Input folder containing subfolders
output_directory = r'C:\Users\3452268\Documents\Biswarup\test\NA221\output'  # Output folder

# Copy the header template to each subfolder inside the input directory
copy_text_file_to_folders(text_file_path, input_directory, output_directory)

# Process the files and update the corresponding output files
process_files(input_directory, output_directory)
```
